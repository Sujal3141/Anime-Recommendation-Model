```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
  def stem(text):
    y=[]
    for i in text.split():
      y.append(ps.stem(i))
    return " ".join(y)
```

```python
anime['tags']=anime['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(anime['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
anime['tags'][0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similar = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similar[0])),reverse=True,key=lambda x:x[1])
```

```python
def recommend(anime_name):
    anime_index = anime[anime['Title']==anime_name].index[0]
    distances = similar[anime_index]
    anime_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in anime_list:
        print(anime.iloc[i[0]].Title)
```

```python
recommend('Trigun')
```
